```python
import pandas as pd
data = pd.read_csv('2018CrimeStatisticsByCategoryOfCrime.csv')
data
```

```python
data.shape
```

```python
data.columns
```

```python
miss = data.isnull().sum(axis = 1) / data.shape[1]*100
rtd = miss[miss>=30].index
new_data = data.drop(rtd)
print(new_data.shape)
```

```python
new_data.fillna(0, inplace=True)
new_data.describe()
```

```python
new_data.duplicated()
```

```python
print(new_data.to_string())
```

```python
new_data.columns
```

```python
import matplotlib.pyplot as plt
plot =  new_data.columns
plt.hist(plot, bins = 30, color = 'grey', edgecolor = 'black')

```
